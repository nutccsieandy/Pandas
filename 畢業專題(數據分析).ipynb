### 資料分析
#### 1.常用套件:Pandas，Numpy
#### 2.實務案例:分析Excel數據
#### 3.生活用途:財務管理
#### 此次選用之專案為:銷貨管理
#### 資料是一個未被整理的數據，例如:貨幣的匯率，或者學生的成績
#### 透過工具整理輔助分析
#### 常用Excel
#### 先用Excel把資料畫成表格，再透過Python去打簡單的指令看我們想要的結果
#### 簡單的程式取代大量計算

```python
df=pd.read_excel("畢業專題(數據分析).xlsx")
ans=pd.DataFrame(df)#資料正規化
#橫列直行(矩陣)
ans
```

```python
ans.loc[ans['價格']>10000]#撈資料1
```

```python
ans.loc[ans['數量']>5]#撈資料2
```

```python
ans.loc[ans['購買人']=="王大明"]#撈資料3
```

```python
ans.sort_values('價格',ascending=True)#價格低到高
#False代表高到低
```

```python
ans.sort_values('價格',ascending=False)#價格高到低
```

```python
ave=ans['毛利'].mean()#計算平均毛利
ave
```

```python
ans['總毛利']
```

```python
sum=0#總和
for i in range(0,5,1):
  sum=sum+ans['總毛利'][i]
print("總金額",sum)
```
